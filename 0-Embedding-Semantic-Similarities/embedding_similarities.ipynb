### Overview 
In this activity, learners work with text embeddings generated by a Large Language Model 
(LLM) to understand how machines represent meaning numerically. Instead of generating 
natural language responses, the model converts text into high-dimensional vectors, which 
can be mathematically compared to measure semantic similarity. 

This activity focuses on embedding generation, vector inspection, and cosine similarity computation, forming the 
foundation for semantic search and Retrieval-Augmented Generation systems. 

#### Learning Objectives 

- Generate text embeddings using an LLM embedding API. 
- Inspect embedding vectors and understand their dimensional properties. 
- Compute cosine similarity between embedding vectors. 
- Compare semantic similarity across multiple text pairs. 
- Understand why semantic similarity differs from keyword-based matching.

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
```

```python
sentences = [
    # Sports domain 
    "I love playing cricket", 
    "I enjoy outdoor sports", 
    "Football is my favorite game", 
 
    # Finance domain 
    "The stock market is volatile today", 
    "Share prices fluctuate frequently", 
    "Investing requires patience", 
 
    # Technology domain 
    "Artificial intelligence is transforming industries", 
    "AI is changing how businesses operate", 
 
    # Completely unrelated 
    "I cooked pasta for dinner last night"
    "I love to code",
    "I want to fall in love",
    "Give me some food"
]

sentences
```

```python
model = SentenceTransformer("all-MiniLM-L6-V2")
embeddings = model.encode(sentences)
```

```python
embeddings
```

```python
embeddings.shape
```

```python
df = pd.DataFrame(embeddings)
df
```

```python
similarity_matrix = cosine_similarity(embeddings)
similarity_matrix
```

```python
df_similarity = pd.DataFrame(
    similarity_matrix,
    columns=sentences,
    index=sentences
)

df_similarity
```

```python
df_similarity.round(2)
```

```python
query_sentence = "Tell me what you know about love"
query_embeddings = model.encode([query_sentence])

query_similarities = cosine_similarity(query_embeddings, embeddings)
query_similarities
```

```python
pd.DataFrame({ 
    "Sentence": sentences, 
    "Similarity Score": query_similarities [0]
}).sort_values(by="Similarity Score", ascending=False) 
```
