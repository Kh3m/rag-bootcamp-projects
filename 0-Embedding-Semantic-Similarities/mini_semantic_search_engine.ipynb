### Overview 

In this activity, learners build a mini semantic search engine by storing text embeddings 
in a vector database and retrieving relevant documents using similarity-based search. 
The activity demonstrates the complete embedding, storage, retrieval pipeline that 
underpins semantic search, Retrieval-Augmented Generation (RAG), and memory 
systems in Agentic AI. 

#### Learning Objectives 

- Generate embedding for a collection of documents. 
- Store embeddings in a vector database using ChromaDB. 
- Perform similarity-based retrieval using natural language queries. 
- Understand how semantic search differs from keyword-based search. 
- Relate semantic search pipelines to real-world AI systems.

```python
from sentence_transformers import SentenceTransformer 
import chromadb 
from chromadb.config import Settings 
import pandas as pd 
```

```python
documents = [ 
    "Cricket is a popular sport played worldwide.", 
    "Football requires teamwork and physical fitness.", 
    "Many people enjoy watching sports on weekends.", 
    "The stock market fluctuates based on economic conditions.", 
    "Investing requires long-term planning and patience.", 
    "Interest rates affect loan and mortgage costs.", 
    "Artificial intelligence is transforming industries.", 
    "Machine learning models learn from data.", 
    "AI systems can automate repetitive tasks.", 
    "Cooking pasta requires boiling water first.", 
    "Traveling helps people experience new cultures." 
] 

documents
```

```python
embedding_model = SentenceTransformer('all-MiniLM-L6-v2') 
```

```python
chroma_client = chromadb.Client( 
    Settings(anonymized_telemetry=False) 
) 

collection = chroma_client.create_collection(name="semantic_search_demo")
```

```python
collection.add( 
    documents=documents, 
    ids=[f"doc_{i}" for i in range(len(documents))] 
) 
 
collection.count() 
```

```python
def semantic_search(query, top_k=2): 
    results = collection.query( 
        query_texts=[query], 
        n_results=top_k 
    ) 
    
    return pd.DataFrame({ 
        "Document": results['documents'][0], 
        "Distance": results['distances'][0] 
    }) 
```

```python
semantic_search("How is AI used in businesses?")
```

```python
semantic_search("How do people invest money wisely?", top_k=5).sort_values(ascending="des")
```
