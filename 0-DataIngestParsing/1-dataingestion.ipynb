### Understanding Document Structure In Langchain

It basically consists of page content which holds the main content of the file and metadata that gives additional information about the file.

#### Why metadata matters:
📝 Metadata is crucial for:
- Filtering search results
- Tracking document sources
- Providing context in responses
- Debugging and auditing

```python
document = Document(
    page_content="Ya'Allah please bless my career and help me to gain a fully funded scholarship in the US or Canada",
    metadata={
        "source": "school.txt",
        "created_at": "06-29-2026"
    }
)
document
```

```python
print(f"Document: {document.page_content}")
print("="*100)
print(f"Metadata: {document.metadata}")
print("="*100)
print(f"Type of Document: {type(document)}")
```

### Text Files (.txt) - The Simplest Case {#2-text-files}

```python
import os
```

```python
os.makedirs("data/text_files", exist_ok=True)
```

```python
text_files = {
    "data/text_files/intro_to_python.txt": """I love to code so we have python here.
    """,
    "data/text_files/intro_to_machine_learning.txt": """Machine learning is a nice to have.
    """
}

for path, content in text_files.items():
    with open(path, "w", encoding="utf-8") as f:
        f.write(content)
        print(f)
```

```python
from langchain_community.document_loaders import TextLoader
```

```python
text_loader = TextLoader("data/text_files/intro_to_python.txt", encoding="utf-8")
document = text_loader.load()
print(f"Document: {document}")
print(f"Metadata: {document[0].metadata}")
print(f"Type of Document: {type(document[0])}")
print(f"Content: {document[0].page_content}")
```

### DirectoryLoader- Multiple Text Files

#### 📊 Analysis
DirectoryLoader Characteristics:

✅ Advantages:
- Loads multiple files at once
- Supports glob patterns
- Progress tracking
- Recursive directory scanning

❌ Disadvantages:
- All files must be same type
- Limited error handling per file
- Can be memory intensive for large directories

```python
from langchain_community.document_loaders import DirectoryLoader
```

```python
dir_loader = DirectoryLoader(
    "data/text_files", 
    loader_cls=TextLoader, 
    glob="**/*.txt", ## Pattern to match .txt files 
    show_progress=True, 
    loader_kwargs={"encoding": "utf-8"}
)

directories = dir_loader.load()
directories
```
