### Understanding Document Structure In Langchain

It basically consists of page content which holds the main content of the file and metadata that gives additional information about the file.

#### Why metadata matters:
📝 Metadata is crucial for:
- Filtering search results
- Tracking document sources
- Providing context in responses
- Debugging and auditing

```python
document = Document(
    page_content="Ya'Allah please bless my career and help me to gain a fully funded scholarship in the US or Canada",
    metadata={
        "source": "school.txt",
        "created_at": "06-29-2026"
    }
)
document
```

```python
print(f"Document: {document.page_content}")
print("="*100)
print(f"Metadata: {document.metadata}")
print("="*100)
print(f"Type of Document: {type(document)}")
```

### Text Files (.txt) - The Simplest Case {#2-text-files}

```python
import os
```

```python
os.makedirs("data/text_files", exist_ok=True)
```

```python
text_files = {
    "data/text_files/intro_to_python.txt": """I love to code so we have python here.
    """,
    "data/text_files/intro_to_machine_learning.txt": """Machine learning is a nice to have.
    """
}

for path, content in text_files.items():
    with open(path, "w", encoding="utf-8") as f:
        f.write(content)
        print(f)
```

```python
from langchain_community.document_loaders import TextLoader
```

```python
text_loader = TextLoader("data/text_files/intro_to_python.txt", encoding="utf-8")
document = text_loader.load()
print(f"Document: {document}")
print(f"Metadata: {document[0].metadata}")
print(f"Type of Document: {type(document[0])}")
print(f"Content: {document[0].page_content}")
```

### DirectoryLoader- Multiple Text Files

#### 📊 Analysis
DirectoryLoader Characteristics:

✅ Advantages:
- Loads multiple files at once
- Supports glob patterns
- Progress tracking
- Recursive directory scanning

❌ Disadvantages:
- All files must be same type
- Limited error handling per file
- Can be memory intensive for large directories

```python
from langchain_community.document_loaders import DirectoryLoader
```

```python
dir_loader = DirectoryLoader(
    "data/text_files", 
    loader_cls=TextLoader, 
    glob="**/*.txt", ## Pattern to match .txt files 
    show_progress=True, 
    loader_kwargs={"encoding": "utf-8"}
)

documents = dir_loader.load()
documents
```

### Text Splitting Statergies


📊 Text Splitting Methods Comparison:

##### CharacterTextSplitter:
Use when: Text has clear delimiters

Adavantages
- ✅ Simple and predictable
- ✅ Good for structured text

Disadavantages
- ❌ May break mid-sentence


##### RecursiveCharacterTextSplitter:
Use when: Default choice for most texts

Advantages
- ✅ Respects text structure
- ✅ Tries multiple separators
- ✅ Best general-purpose splitter

Disvantages
- ❌ Slightly more complex
  

##### TokenTextSplitter:
Use when: Working with token-limited models

Advantages:
- ✅ Respects model token limits
- ✅ More accurate for embeddings

Disadvantages
- ❌ Slower than character-based
  

```python
from langchain_text_splitters import (
    CharacterTextSplitter,
    RecursiveCharacterTextSplitter,
    TokenTextSplitter
)
```

```python
document
```

```python
char_splitter = CharacterTextSplitter(
    separator=" ",      # Split on blank space
    chunk_size=20,      # Max chunk size in characters
    chunk_overlap=5,    # Overlap between chunks
    length_function=len # How to measure chunk size
)

text_chunks = char_splitter.split_text(document[0].page_content)
text_chunks
```

```python
char_splitter.split_documents(document)
```

```python
recursive_splitter = RecursiveCharacterTextSplitter(
    separators=["\n\n", "\n", " ", ""], 
    chunk_size=100, 
    chunk_overlap=10
)

for doc in documents:
    doc_splits = recursive_splitter.split_text(doc.page_content)
    print(doc_splits)
```

```python
token_splitter = TokenTextSplitter(
    chunk_size=50,  # Size in tokens (not characters)
    chunk_overlap=2,
)
```

```python
token_chunks = token_splitter.split_text(documents[2].page_content)
len(token_chunks)
print(token_chunks[0])
print("-"*50)
print(token_chunks[1])
print("-"*50)
print(token_chunks[2])
```
